```python
from fancyimpute import KNN
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline

cols = ['Application mode', 'Course', 'Previous qualification', "Mother's qualification", "Father's qualification", 'Tuition fees up to date', 
        "Mother's occupation", "Father's occupation", 'Gender', 'Scholarship holder', 'Age at enrollment', 'Curricular units 1st sem (approved)', 
        'Curricular units 2nd sem (approved)', 'Target']

df = new_df[cols]

df = df[df['Target'] != 'Enrolled']

df = df.replace({'Target': {'Dropout': 0, 'Graduate': 1}})

df['Tuition fees up to date'] = KNN(k=3).fit_transform(df[['Tuition fees up to date']])
df['Scholarship holder'] = KNN(k=3).fit_transform(df[['Scholarship holder']])

impute_cat_cols = ['Application mode', 'Course', 'Previous qualification']

for col in impute_cat_cols:
    features = df.dropna().drop(col, axis=1)
    target = df.dropna()[col]

    imputer = make_pipeline(SimpleImputer(strategy='most_frequent'), DecisionTreeClassifier())
    imputer.fit(features, target)

    df[col] = df.apply(lambda row: row[col] if not pd.isna(row[col]) else imputer.predict([row.drop(col)])[0], axis=1)

impute_unit_cols = ['Curricular units 1st sem (approved)', 'Curricular units 2nd sem (approved)']

for col in impute_unit_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)

cols = ['Application mode', 'Course', 'Previous qualification', "Mother's qualification", "Father's qualification", 'Tuition fees up to date', 
        "Mother's occupation", "Father's occupation", 'Gender', 'Scholarship holder', 'Age at enrollment', 'Target']
df[cols] = df[cols].astype('int32')
```

```python
X = df.drop('Target', axis=1)
y = df['Target']

print(X['Mother\'s occupation'].dtypes)

print(X.head())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

label_encoder = LabelEncoder()
X['Mother\'s occupation'] = label_encoder.fit_transform(X['Mother\'s occupation'])
X['Father\'s occupation'] = label_encoder.fit_transform(X['Father\'s occupation'])
X['Mother\'s qualification'] = label_encoder.fit_transform(X['Mother\'s qualification'])
X['Father\'s qualification'] = label_encoder.fit_transform(X['Father\'s qualification'])

print(X.head())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, SimpleRNN, LeakyReLU

fnn_model = Sequential()
fnn_model.add(Dense(128, input_dim=X_train.shape[1], activation=LeakyReLU(alpha=0.3)))
fnn_model.add(Dense(64, activation=LeakyReLU(alpha=0.2)))
fnn_model.add(Dense(32, activation=LeakyReLU(alpha=0.1)))
fnn_model.add(Dense(16, activation=LeakyReLU(alpha=0.01)))
fnn_model.add(Dense(1, activation='sigmoid'))
fnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_fnn = fnn_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

fnn_model.save('fnn_model.h5')
```

```python
X_cnn = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

cnn_model = Sequential()
cnn_model.add(Conv1D(64, kernel_size=3, activation=LeakyReLU(alpha=0.3), input_shape=(X_cnn.shape[1], 1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Conv1D(128, kernel_size=3, activation=LeakyReLU(alpha=0.2)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(64, activation=LeakyReLU(alpha=0.1)))
cnn_model.add(Dense(32, activation=LeakyReLU(alpha=0.01)))
cnn_model.add(Dense(1, activation='sigmoid'))
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_cnn = cnn_model.fit(X_cnn, y, epochs=100, batch_size=32, validation_split=0.2)
```

```python
X_rnn = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))

rnn_model = Sequential()
rnn_model.add(SimpleRNN(64, input_shape=(X_rnn.shape[1], X_rnn.shape[2]), return_sequences=True))
rnn_model.add(SimpleRNN(32, return_sequences=True))
rnn_model.add(SimpleRNN(16))
rnn_model.add(Dense(1, activation='sigmoid'))
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_rnn = rnn_model.fit(X_rnn, y, epochs=100, batch_size=32, validation_split=0.2)
```

```python
plt.subplot(1, 2, 1)
plt.plot(history_fnn.history['loss'], label='FNN Train')
plt.plot(history_fnn.history['val_loss'], label='FNN Validation')
plt.plot(history_cnn.history['loss'], label='CNN Train')
plt.plot(history_cnn.history['val_loss'], label='CNN Validation')
plt.plot(history_rnn.history['loss'], label='RNN Train')
plt.plot(history_rnn.history['val_loss'], label='RNN Validation')

plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.plot(history_fnn.history['accuracy'], label='FNN Train')
plt.plot(history_fnn.history['val_accuracy'], label='FNN Validation')
plt.plot(history_cnn.history['accuracy'], label='CNN Train')
plt.plot(history_cnn.history['val_accuracy'], label='CNN Validation')
plt.plot(history_rnn.history['accuracy'], label='RNN Train')
plt.plot(history_rnn.history['val_accuracy'], label='RNN Validation')

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
